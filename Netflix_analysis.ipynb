```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import the csv file from desktop

```python
df=pd.read_csv(r'C:\Users\Saumya\Downloads\archive.zip')
df
```

### Check for empty values and duplicated values and name the null values to unknown and delete the rows with empty date.

```python
df.isnull().sum()
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Unknown')
df['duration'] = df['duration'].fillna('Unknown')
df
```

```python
df.dropna(subset=['date_added'], inplace=True)
```

### Change the datatype of date from string to datetime format

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.dtypes
```

### delete unnecessary column description

```python
df.drop('description', axis=1, inplace=True)
df
```

### extract year and month 

```python
df['year'] = df['date_added'].dt.year
df['month_name'] = df['date_added'].dt.month_name()
df
```

### 1. Total number of movies vs TV shows

```python
content_count= df['type'].value_counts()
```

```python
plt.figure(figsize=(12,4.5))
plt.pie(content_count,
        labels=content_count.index,
        autopct='%1.1f%%',
        startangle=90, colors=['orange','red'])

plt.title('Movies vs TV Shows Distribution on Netflix')
plt.show()
```

### 2. Content distribution by type and year of release(visualization)

```python
plt.figure(figsize=(12,4.5))
a= sns.countplot(data=df, x='type', hue="year", palette="Set1")
plt.title('Distribution based on Type')
for container in a.containers:
     a.bar_label(container)
plt.xlabel('Type')
plt.show
```

### 5. Top 10 content producing countries

### 6. Most common genres

```python
df['listed_in'] = df['listed_in'].str.split(',')
genre_df = df.explode('listed_in')
genre_df['listed_in'] = genre_df['listed_in'].str.strip()
genre_df['listed_in'].value_counts().head(10)
```

### Explanation to the above code
Firstly str.split() was used to split the text in the column 'listed_in' based on the separator ','. After str.split() the values stored as(example) "International TV Shows, TV Dramas, TV Mysteries" now appeared as a list, now each value could be accessed using an index.

Now the explode() function used on the column 'listed_in' converted the listed values into multiple rows. 

Code where str.strip() is used in the next line denotes removal of extra spaces.

The next part is where we count the frequency of each genre by using value_counts() that counts how many times each value appears..then finally we selected top 10 genre which were repeated the most.



### 7. Duration Analysis

```python
movies= df[df['type'] == 'Movie']
movies['duration'] = movies['duration'].astype(str).str.extract(r'(\d+)')
movies['duration'] = pd.to_numeric(movies['duration'], errors='coerce')
tv= df[df['type'] == 'TV Show']
tv['duration'] = tv['duration'].astype(str).str.extract(r'(\d+)')
tv['duration'] = pd.to_numeric(tv['duration'], errors='coerce')
```

```python
plt.figure(figsize=(12,4.5))
sns.boxplot(data=movies, x='duration', color='seagreen')
plt.title("Movie Duration (Minutes)")
plt.show()
```

```python
plt.figure(figsize=(12,4.5))
sns.boxplot(data=tv, x='duration', color='brown')
plt.title("TV Show Duration (Seasons)")
plt.show()
```

### Explanation to the above codes
The movies and tv store a separate dataframe where 'type' column is present based on condition. Basically they separate the movies and tv shows category.
Then we plot the boxplot separately for movies (duration in minutes) and tv shows (duration in seasons) where data is not df but the movies and tv which were inititally used to store the conditional column.

Boxplot helps us answer:
“What is typical in the given data?”
“How much variation can be seen?”
“Are there extreme values?”

The movies boxplot gave us insights that - Most movies lie between 80–120 minutes, with some long-duration outliers above 150 minutes.
while for the tv shows - Most shows have 1–3 seasons, with a few long-running series as outliers.

### 9. Year Growth Rate

```python
yearly_trend.pct_change()
```

This particular code calculates year-over-year percentage change. Basically answers “How much did the number of shows/movies change compared to the previous year?

NaN in 2008 denotes that there is no previous year to compare hence NaN.


```python
plt.figure(figsize=(12,4.5))
(yearly_trend.pct_change() * 100).plot(kind='line', marker='o', color='orange')
plt.title("Year-over-Year Growth (%)")
plt.ylabel("Growth %")
plt.show()
```

The insight we can conclude from this information is -
Netflix content experienced rapid expansion between 2011–2016, with peak growth exceeding 400% in some years, followed by stabilization and slight decline post-2019.
